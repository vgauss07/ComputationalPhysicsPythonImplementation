## Taylor Series Expansion

A continous function can be expressed as a polynomial

Equation for Taylor Series:
$$
\\ f(x)=\sum_{n=0}^\infty \frac{1}{n!}f^{(n)}(x_0)\,(x-x_0)^n\\
$$
where,
$$
\\(f^{(n)})\\
$$
is the nth derivative

## Taylor's Series Expansion of Exponential

```python
def taylorExp(x, x0, n_max):
    """

    :param x: Argument
    :param x0: Argument at which the derivatives will be calculated
    :param n_max: n at which the series will terminate
    :return:
    """
    t = 0
    for n in range(n_max+1):
        t = t + np.exp(x0)*(x-x0)**n / np.math.factorial(n)
    return t
```

```python
taylorExp(1, 0, 30)
```

```python
np.exp(1)
```

## Accuracy Check

```python
np.exp(1) - taylorExp(1, 0, 30)
```

## Visualize Convergence of the Series Expansion of Exponential, where n = 30

```python
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([-5,100])


xList = np.linspace(-5, 5, 101)
plt.scatter(xList, np.exp(xList))

plt.plot(xList, taylorExp(xList, 0, n_max=30), 'red')
plt.plot(xList, taylorExp(xList, -3, n_max=30), 'green')
plt.plot(xList, taylorExp(xList, 2, n_max=30), 'orange')
```

## Taylor's Series Expansion of Sin Function

sin function at
$$
\\ x_0 = 0 \\
$$

$$
\\ ( f(0) = f''(0) = f^{(4)}(0) = \dots = 0 ) \\
\\ ( f'(0) = f^{(5)}(0) = f^{(9)}(0) = \dots = 1 ) \\
\\ (f'''(0) = f^{(7)}(0) = f^{(11)}(0) = \dots = -1 ) \\
\\ ( \sin(x) = x - \frac{1}{3!} + \frac{1}{5!}x^5 - \frac{1}{7!}x^7 \pm \dots = \sum_{n=0}^\infty \frac{(-1)^n}{(2n+1!}x^{2n+1}) \\

$$


```python
def taylorSin(x, n_max):
    """
    :param x: Argument
    :param n_max: n at which the series will terminate
    :return:
    """
    t = 0  # initialize variable
    for n in range(n_max+1):
        t = t + (-1)**n * x**(2*n+1) /np.math.factorial(2*n + 1)
    return t
```

## Accuracy Check

```python
np.sin(10.5) - taylorSin(10.5, 50)
```

## Test Convergence of the Series Expansion of Sin

```python
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([-2,2])


xList = np.linspace(-10, 10, 101)
plt.scatter(xList, np.sin(xList))

plt.plot(xList, taylorSin(xList, n_max=3), 'red')
plt.plot(xList, taylorSin(xList, n_max=6), 'green')
plt.plot(xList, taylorSin(xList, n_max=9), 'orange')
plt.plot(xList, taylorSin(xList, n_max=12), 'black')
```
